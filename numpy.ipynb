```python
#dot product
r=np.array([[1,2],[3,4]])
t=np.array([[6,7],[8,9]])
c=r.dot(t)
c
```

```python
#concatenate
v=np.array([1,2,3,4])
x=np.array([7,8,9,7])
mn = np.array([[1,2,3],[4,5,6]])
np.concatenate((v,x))
np.append(v,100)
np.append(mn, [[10],[20]],axis=1)
```

```python
#append and insert
op = np.array([1,2,3,4,5,6])
np.insert(op,1,100)
np.insert(op,(1,3),4)
```

```python
#insert 2 dim
de = np.array([[1,2,3],[4,5,6]])
np.insert(de,1,100,axis=0)
np.insert(de,1,[99,89,78],axis=0)
```

```python
#delete member in array
qs = np.arange(1,10)
np.delete(qs,2)
gh = np.array([[2,3,4],[3,4,5]])
np.delete(gh,1,axis=0)
```
